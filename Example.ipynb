# Example of IPython notebook in Markdown

Plot the curve $y=f(x)$, where

$$
f(x) = e^{-x}\sin (2\pi x), \quad x\in [0,4]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline 

x = np.linspace(0, 4, 101)
y = np.exp(-x)*np.sin(2*math.pi*x)
plt.plot(x, y, 'b-')
plt.xlabel('x')
plt.ylabel('y')
```
